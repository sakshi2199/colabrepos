```
#in case of using laptop/pc, pip install matplotlib and seaborn
#!pip install matplotlib seaborn #in colab 
import matplotlib.pyplot as plt
x=[2,5,4,9]
y=[3,8,3,6]
```

```
plt.xlabel('distance') # for labeling x axis
plt.ylabel('time') #for labeling y axis
plt.scatter(x,y) #to show two dots only
#plt.scatter(y,x)
plt.plot(x,y, label="railway track")
plt.plot(x,y) #to plot line
plt.bar(x,y, color='violet', label="finance growth" )#show bar graph
plt.bar(y,x)
plt.legend() #to show labels of plot area
plt.grid(c='grey') # for grid and color
plt.plot() #to plot the graph
```

```
#cricket score graph
players=["Dhoni", "Rohit", "Virat"]
runs =[250, 113, 230]
plt.xlabel("players")
plt.ylabel("runs")
#plt.bar(players, runs, label="2020 starting match")
plt.pie(runs, labels=players, autopct='%1.1f%%', explode=(0,0.2,0), shadow=True)
plt.legend()
plt.show()
```
